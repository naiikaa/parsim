```python
import numpy as np
import matplotlib.pyplot as plt
import deeptrack as dt
import tqdm
import cv2
from multiprocessing import Pool


WIDTH, HEIGHT= 1920, 1080
IMAGE_SIZE = 1920

particles = dt.MieSphere(
        refractive_index = 1.44, 
        position = lambda: (
            np.random.random()*HEIGHT,
            np.random.random()*WIDTH,
        ),
        radius = 1.4*1e-6,
        position_unit="pixel",
        z = lambda: -10+np.random.random()*20,
        L=10,
        intensity=10,
    )

spectrum = np.linspace(300e-9,700e-9, 10)


microscopes = [dt.Brightfield(NA=.7,
                                wavelength=wl,
                                magnification=50,
                                output_region=(0, 0, HEIGHT, WIDTH),
                                refractive_index_medium=1,
                                resolution = 14e-6,
                                )
                for wl in spectrum]

noise = dt.Gaussian(sigma=.1)

particle_number = lambda: np.random.randint(99,125+1)

sample = (particles^particle_number)
image = sum([microscope(sample) for microscope in microscopes])
augmented_image = (image >>noise >> dt.NormalizeMinMax())
```

```python
for _ in range(1):
    augmented_image.update()
    plt.figure(figsize=(10,6),dpi=300)
    augmented_image.plot(cmap='gray')
    plt.show()
```
